### Standard Scalar

```python
def StandardScaler(col):
  s = col.std()
  mean = col.mean()
  col = col.apply(lambda x: (x - mean)/s)
  return col
```

```python
#implement Kmeans from scratch
class Kmeans:

    def __init__(self,train_data):
        self.train_data = train_data.astype('int32')

    def init_clusters(self,K):
        length_dataset = len(self.train_data)
        indexes = random.sample(range(length_dataset),k=K) #Do not hard code thiss
        self.clusters = self.train_data[indexes]
    
    def init_clusters_random(self,K):
        length_dataset = len(self.train_data)
        random_cluster_id = []
        for x in range (0, length_dataset):
            random_cluster_id.append(random.randint(0, K-1))
        random_cluster_id = np.array(random_cluster_id)
        self.clusters = []
        for i in range(K):
            indices = np.where(random_cluster_id == i)[0].tolist()
            sample_data = np.sum(self.train_data[indices],axis=0)/len(indices)
            self.clusters.append(sample_data)
        self.clusters = np.array(self.clusters)
        
    def find_distances(self,distance_measure="euclidean"):
        if(distance_measure == "euclidean"):
            dist = -2*np.dot(self.train_data,self.clusters.T) + np.sum(self.clusters**2,axis=1) + np.sum(self.train_data**2,axis=1)[:,np.newaxis]
            # dist = np.where(dist<0,0,dist)
            # print(np.sqrt(dist).shape)
            return np.sqrt(dist)
        elif(distance_measure == "manhattan"): 
            return np.abs(self.train_data[:,None] - self.clusters).sum(-1)

    def fit_data(self,distance_measure):
        for i in range(10):
            cluster_distance_matrix = self.find_distances(distance_measure)
            # print(cluster_distance_matrix.shape)
            nearest_cluster_matrix = np.argmin(cluster_distance_matrix,axis=1) 
            # print(np.unique(nearest_cluster_matrix))
            for i in range(len(self.clusters)):
                indices = np.where(nearest_cluster_matrix==i)[0].tolist()
                # print(indices)
                self.clusters[i] = self.train_data[indices].mean(axis=0)
        self.cluster_distance_matrix = self.find_distances()
        self.nearest_cluster_matrix = np.argmin(self.cluster_distance_matrix,axis=1)
     
    def predict(self,distance_measure,test_data):
        cluster_distance_matrix = np.sqrt(-2*np.dot(test_data,self.clusters.T) + np.sum(self.clusters**2,axis=1) + np.sum(test_data**2,axis=1)[:,np.newaxis])
        nearest_cluster_matrix = np.argmin(cluster_distance_matrix,axis=1)
        return nearest_cluster_matrix
```

```python
def silhouette(train_data,K,dist_measure):
    mean_distance_matrix = np.array([0 for i in range(len(train_data))])
    kmean = Kmeans(train_data)
    kmean.init_clusters(K)
    kmean.fit_data(dist_measure)
    for i in range(K):
        indices = np.where(kmean.nearest_cluster_matrix == i)[0].tolist()
        X = train_data[indices]
        dist = -2*np.dot(X,X.T)+ np.sum(X**2,axis=1)+np.sum(X**2,axis=1)[:,np.newaxis]
        dist = np.where(dist<0,0,dist)
        matrix = np.sum(np.sqrt(dist),axis=1)/(len(indices)-1)
        mean_distance_matrix[indices] = matrix
    C_second_nearest_cluster = np.argsort(kmean.cluster_distance_matrix,axis=1)[:,1]
    mean_dissimilarity_matrix = np.array([0 for i in range(len(train_data))])
    for i in range(K):
        indices_x = np.where(C_second_nearest_cluster == i)[0].tolist()
        X = train_data[indices_x]
        indices = np.where(kmean.nearest_cluster_matrix == i)[0].tolist()
        Y = train_data[indices]
        dist = -2*np.dot(X,Y.T)+ np.sum(Y**2,axis=1)+np.sum(X**2,axis=1)[:,np.newaxis]
        dist = np.where(dist<0,0,dist)
        matrix = np.sum(np.sqrt(dist),axis=1)/(len(indices))
        mean_dissimilarity_matrix[indices_x] = matrix
    # print(mean_dissimilarity_matrix, mean_distance_matrix)
    silhoutte_scores = (mean_dissimilarity_matrix - mean_distance_matrix)/np.maximum(mean_dissimilarity_matrix,mean_distance_matrix)
    return np.sum(silhoutte_scores)/len(silhoutte_scores)
```

```python

def elbow(K,distance_measure,train_data):
    cost_function_array = []
    for k in range(3,K):
        total_sum = 0
        kmean = Kmeans(train_data)
        kmean.init_clusters(k)
        kmean.fit_data(distance_measure)
        for i in range(len(kmean.clusters)):
            indices = np.where(kmean.nearest_cluster_matrix==i)[0].tolist()
            total_sum += np.sum((kmean.train_data[indices] - kmean.clusters[i])**2)
        cost_function_array.append(total_sum)
    K = [i for i in range(3,K)]
    plt.plot(K, cost_function_array, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel('Distortion')
    plt.title('The Elbow Method using Distortion')
    plt.show()
    return cost_function_array
```

### Silhoutte method

```python
sil_score = []
for i in range(3,15):
    ss = silhouette(PCA_dataset.to_numpy(),i,'euclidean')
    sil_score.append(ss)
```

```python
K = [i for i in range(3,15)]
plt.plot(K, sil_score, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Silhouette Method using Distortion')
plt.show()
```
